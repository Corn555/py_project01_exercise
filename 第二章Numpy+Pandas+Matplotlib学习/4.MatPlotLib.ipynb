```python
import matplotlib.pyplot as plt
import numpy as np

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales = [120, 150, 170, 160, 200, 230]

profit = [20, 30, 35, 32, 45, 55]
```

```python
# 折线图
plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker="o", color="red", linestyle="--", label="sales")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.grid(True)
plt.ylim(100,250)
plt.legend()
plt.show()
```

```python
#柱状图
plt.figure(figsize=(8, 5))
plt.bar(months, profit, color="green", label="profit", width=0.5)
plt.title("Sales Bar Chart")
plt.xlabel("Months")
plt.ylabel("Profit")
plt.legend()
plt.show()
```

```python
#散点图
hours = [1,2,3,4,5,6,7]
scores = [50,55,65,70,80,90,95]
plt.figure(figsize=(8, 5))
plt.scatter(hours, scores, marker="o", color="red", label="scores", s=100)
```

```python
# 折线图
plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker="o", color="blue", linestyle="--", label="sales")
plt.plot(months, profit, marker="o", color="green", label="profit")

plt.grid(True)
plt.legend()
plt.show()
```

```python
# 折线图
plt.figure(figsize=(9, 5))
plt.plot(months, sales, marker="o", color="blue", linestyle="--", label="sales")
plt.plot(months, profit, marker="o", color="red", label="profit")
plt.title("Sales & Profit")
plt.xlabel("Months")
plt.ylabel("Amount")
plt.grid(True)
plt.legend()
plt.show()
```
