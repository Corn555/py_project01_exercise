```python
import numpy as np

"""
某周天气最高温：28，30，29，31，32，30，29
求平均气温，最高最低气温
求超过30°的天数
"""
```

```python
temp = np.array([28,30,29,31,32,30,29])
print(np.max(temp))
print(np.min(temp))
print(np.mean(temp))
print(len(temp[temp>30]))
```

```python
'''
某班级5个学生的成绩L:85,90,78,92,88
计算成绩的平均分，中位数和标准差
将成绩转换成十分制（假设满分10）
'''
```

```python
score = np.array([85,90,78,92,88])
print(np.mean(score))
print(np.median(score))
print(np.std(score))#标准差
```

```python
'''
矩阵A = [[1,2],[3,4]]  B=[[5,6],[7,8]]
计算A+B A*B
A,B的矩阵乘法
'''
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print(A+B)
print(A*B)
print(A@B)
```

```python
'''
生成一个（3，4）的随机整数数组范围0-10
计算每列max，每行min
将数组中的所有奇数替换成-1
'''
```

```python
import numpy as np

np.random.seed(0)
arr = np.random.randint(0,10,(3,4))
print("原始数组：")
print(arr)
print("每列最大值：")
print(arr.max(axis=0))
print("每行最小值：")
print(arr.min(axis=1))
print("奇数替换为-1：")
print(np.where(arr % 2 == 1,-1,arr))
```

```python
'''
创建一个1-12一维数组，并转换成（3，4）的二维数组
计算每行的和，每列的平均值
将数组展平为一维数组
'''
```

```python
arr1 = np.arange(1,13)
arr2 = arr1.reshape(3,4)
print('arr1:',arr1)
print('arr2:')
print(arr2)
print(arr2.sum(axis=1))
print(arr2.mean(axis=0))
arr3 = arr2.reshape(12)
print('arr3:',arr3)
```

```python
np.random.seed(0)
arr = np.random.randint(0,20,(5,5))
print('arr:')
print(arr)
print('arr1:')
print(arr[arr>10])
print('arr2:')
print(np.where(arr>10,0,arr))
```
