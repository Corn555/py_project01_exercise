```python
#Part5
# #55
# 按 species 分组，计算平均体重。
#
df.groupby('species')['body_mass_g'].mean()
# #56
# 按 species 分组，计算最大体重。
#
df.groupby('species')['body_mass_g'].max()
# #57
# 按 species 分组，计算最小体重。
#
df.groupby('species')['body_mass_g'].min()
# #58
# 按 sex 分组，计算平均体重。
df.groupby('sex')['body_mass_g'].mean()
```

```python
# Part6 GroupBy

# #59
# 按 island 分组，计算平均体重。
#df.groupby('island')['body_mass_g'].mean()

# #60
# 按 species 分组，统计每种企鹅有多少只。
# （提示：使用 count()）
df.groupby('species').count()


# #61
# 按 species 分组，计算平均嘴长。
df.groupby('species')['bill_length_mm'].mean()

# #62
# 按 species 分组，计算平均鳍长。
df.groupby('species')['bill_depth_mm'].mean()

# #63
# 按 species 分组，同时统计：
# - 平均体重
# - 最大体重
# - 最小体重
# （提示：使用 agg()）
df.groupby('species')['body_mass_g'].agg(['mean', 'max', 'min'])

# #64
# 按 sex 分组，同时统计：
# - 平均体重
# - 平均嘴长
df.groupby('sex')['body_mass_g','bill_length_mm'].mean()

# #65
# 按 island 分组，同时统计：
# - 平均体重
# - 平均鳍长
df.groupby('island')['body_mass_g','bill_depth_mm'].mean()

# #66
# 按 species 分组，统计：
# - body_mass_g 平均值
# - bill_length_mm 平均值
# - flipper_length_mm 平均值
df.groupby('species')['body_mass_g','bill_length_mm','flipper_length_mm'].mean()

# #67
# 按 species 分组，统计：
# body_mass_g 的
# - 平均值
# - 最大值
# - 最小值
# （提示：使用 agg()）
df.groupby('species')['body_mass_g'].agg(['mean', 'max', 'min'])

# #68
# 按 species 和 sex 两列一起分组，
# 计算平均体重。
df.groupby('species','sex')['body_mass_g'].mean()

# #69
# 按 island 和 species 两列一起分组，
# 统计每组企鹅数量。
df.groupby('species','island')['species'].count()

# #70（Boss）
# 输出下面这样的统计表：
#
# species
# 平均体重
# 最大体重
# 最小体重
# 平均嘴长
# 平均鳍长
# 数量
df.groupby('species').agg({
    'body_mass_g': ['mean', 'max', 'min'],
    'bill_length_mm':'mean',
    'flipper_length_mm':'mean',
    'spexies':'count',
})

```
