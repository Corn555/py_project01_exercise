```python
import numpy as np
import pandas as pd
'''
计算每位学生的总分和平均分
找出数学大于90或英语高于85的学生
按总分高到低排序，并输出前三个学生
'''
Student_score = pd.DataFrame(
    {
        'name': ['zhang','li','wang','zhao','qian'],
        'math': [85,92,78,88,95],
        'english': [90,88,85,92,80],
        'physics': [75,80,88,85,90]
    }
)
Student_score['total'] = Student_score['math'] + Student_score['english'] + Student_score['physics']
print(Student_score['total'])

Student_score['avg'] = Student_score['total'] / 3


print(Student_score[(Student_score['math'] > 90) | (Student_score['english'] > 85)])

print(Student_score.sort_values(by='total', ascending=False).head(3))
```

```python
import pandas as pd
'''
计算每种产品的总销售额
找出销售额最高的产品
按照销售额从高到低并且输出所有产品信息
'''
sales_data = pd.DataFrame(
    {
        'name': ['A','B','C','D'],
        'price': [100,150,200,120],
        'num': [50,30,20,40]
    }
)
sales_data['total'] = sales_data['price'] * sales_data['num']

print(sales_data.sort_values(by='total', ascending=False).head(1))

print(sales_data.sort_values(by='total', ascending=False))
```

```python
import pandas as pd

data = pd.DataFrame(
    {
        'store' : ["一号店","一号店","二号店","二号店","一号店","二号店","三号店","三号店"],
        'product' : ["牛奶","面包","牛奶","零食","零食","面包","牛奶","零食"],
        'price' : [80, 15, 80, 25, 25, 15, 80, 25],
        'sales_num' : [70, 120, 65, 110, 90, 130, 55, 140],
        'month' : [3, 3, 3, 4, 3, 4, 4, 4]
    }
)
#新增列 turnover 单品销售额 = 单价 × 销量
data['turnover'] = data['price'] * data['sales_num']

#筛选条件：3 月份 且 销售额大于 5000 的所有记录并打印
print(data[ (data['month'] == 3) & (data['turnover'] > 5000) ])

#按门店分组，计算每家门店总销售额，输出分组结果
store_total = data.groupby("store")["turnover"].sum()
print("\n各门店总销售额：")
print(store_total)

#找出门店总销售额排名第一的门店名称及对应总销售额
max_data_name = store_total.sort_values(ascending=False).head(1).idxmax()
max_data_sale = store_total.sort_values(ascending=False).head(1).max()
print(f'总销售额排名第一的门店名称:{max_data_name},总销售额:{max_data_sale}')


#原始全表按销售额turnover降序排序，输出前 4 条数据
print(data.sort_values(by='turnover', ascending=False).head(4))
```
