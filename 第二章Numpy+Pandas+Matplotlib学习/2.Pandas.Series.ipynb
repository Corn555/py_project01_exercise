```python
from operator import index

'''
创建十个学生数学成绩，范围在50-100
计算mean,max,min,并且找出高于mean的学生人数
'''
import pandas as pd
import numpy as np

# 设置随机种子保证结果固定
np.random.seed(0)
# 生成10个50~99的整数成绩（randint左闭右开，上限写100）
s = pd.Series(np.random.randint(50, 100, 10))
print("所有学生数学成绩：")
print(s)

# 计算均值、最大值、最小值
mean_score = s.mean()
max_score = s.max()
min_score = s.min()

print(f"\n平均分mean：{mean_score}")
print(f"最高分max：{max_score}")
print(f"最低分min：{min_score}")

# 筛选高于平均分的成绩，统计人数
above_mean = s[s > mean_score]
count_above = len(above_mean)
print(f"\n高于平均分的学生人数：{count_above}")
print("高于平均分的成绩：")
print(above_mean)
```
